```python
# define function that gets context concreteness score and get scores for words
def get_neighborhood_con(seq_embeds, con_set, abs_set, word_set, start_year=1900, end_year=2000, num_n=10, first_order=False):
    con_scores = collections.defaultdict(lambda : collections.defaultdict(lambda : float('nan')))
    for year in range(start_year, end_year + 1):
        year_embed = seq_embeds.get_embed(year)
        for word in word_set:
            if first_order:
                neighs = year_embed.closest_first_order(word, n=num_n + 1)
            else:
                neighs = year_embed.closest(word, n=num_n + 1)
            neigh_set = set([entry[1] for entry in neighs if entry[1] != word and entry[0] != 0])
            if len(neigh_set) <= 1:
                score = float('nan')
            else:
                score = (len(neigh_set.intersection(con_set))-len(neigh_set.intersection(abs_set)))/float(len(neigh_set))
            con_scores[word][year] = score
    return con_scores
# scores for all words per year
con_scores = get_neighborhood_con(first_orders, set(concrete_cmn), set(abs_cmn), val_set, start_year=1900, first_order=True)
# yearly mean and stds
con_scores_ms = seriesanalysis.get_series_mean_std(con_scores, con_scores.keys())
```

```python
import scipy.sparse as sparse
from sklearn.preprocessing import normalize
import random
def get_entry_density_fast(new_words, seq_embeds, sum_stat, knn_sum_stat, sums, knn_sums, seed=10, num_nn=5, num_rand=10):
    reals = {"avg-cos" : np.empty((len(new_words))), "nn-cos" : np.empty((len(new_words)))}
    counter = 0
    for word, year in new_words.iteritems(): 
        reals["nn-cos"][counter] = knn_sum_stat[word][year]/knn_sums[year-1900]
        reals["avg-cos"][counter] = sum_stat[word][year]/sums[year-1900]
        counter +=1 
    return reals

def get_entry_nn_density_fast(new_words, seq_embeds, seed=10, num_nn=5):
    reals = {"avg-cos" : np.empty(num_nn*(len(new_words))), 
             "nnz" : np.empty((num_nn*len(new_words))), 
             "nn-cos" : np.empty(num_nn*(len(new_words)))}
    counter = 0
    for t_word, year in new_words.iteritems():
        nns = [entry[1] for entry in seq_embeds.get_embed(year).closest(t_word, n=num_nn+1)[1:]]
        for word in nns:
            year_embed = seq_embeds.get_embed(year)
            word_embed = year_embed.represent(word)
            real_dist = np.sort((year_embed.m.dot(word_embed.T).data))
            reals["nn-cos"][counter] = real_dist[-num_nn-1:-2].sum()
            reals["avg-cos"][counter] = real_dist.sum()
            reals["nnz"][counter] = word_embed.nnz
            counter +=1 
    return reals
```

```python
def get_seq_closest(embeds, word, start_year, num_years=10, n=15):
    closest = collections.defaultdict(float)
    for year in range(start_year, start_year + num_years):
        year_closest = embeds.get_embed(year).closest(word, n=100*n)
        for score, neigh in year_closest:
            closest[neigh] += score
    return sorted(closest, key = lambda neigh : closest[neigh], reverse=True)[0:n+1]
```

```python
def get_year_vals(word_year_dict, df, val, year_range=1):
    vals = []
    word_set = set(df["word"].unique())
    df = df.set_index(["word"])
    for word, year in word_year_dict.iteritems():
        if word in word_set:
            if type(df.loc[word]["year"]) == np.float64:
                continue
            v = df.loc[word].loc[df.loc[word]["year"] == year][val].as_matrix()
            if len(v) != 0:
                vals.append(np.mean(v))
    return vals
```

```python
def make_nn_dist_series(word_births, val_func, years=30):
    series = collections.defaultdict(dict)
    for word, birth_year in word_births.iteritems():
        if birth_year > 1970:
            continue
        for i, year in enumerate(range(birth_year, birth_year + years)):
            val = val_func(word, year)
            if val == -1*np.inf:
                series[word][i] = np.nan
                continue
            series[word][i] = val
    return series
knn_func = lambda word, year : log_knn_sums[word][year] - log_knn_sums_med[year - 1900]
knn_s_series = make_nn_dist_series(succ, knn_func)
knn_u_series = make_nn_dist_series(unsucc, knn_func)
sum_func = lambda word, year : np.log(sums[word][year]) - log_sums_med[year - 1900]
sum_s_series = make_nn_dist_series(succ, sum_func)
sum_u_series = make_nn_dist_series(unsucc, sum_func)
freq_func = lambda word, year : np.log(year_freqs[word][year]) - log_freqs_med[year - 1900]
freq_s_series = make_nn_dist_series(succ, freq_func)
freq_u_series = make_nn_dist_series(unsucc, freq_func)
clust_func = lambda word, year : np.log(clusts[word][year]) - log_clusts_med[year - 1900]
clust_s_series = make_nn_dist_series(succ, clust_func)
clust_u_series = make_nn_dist_series(unsucc, clust_func)
```

```python
con_func = lambda word, year : (con_scores[word][year] - con_scores_ms[0][year-1901])/con_scores_ms[1][year-1901]
con_s_series = [val for vals in make_nn_dist_series(succ, con_func, years=1).values() for val in vals.values() if not np.isnan(val)]
con_u_series = [val for vals in make_nn_dist_series(unsucc, con_func, years=1).values() for val in vals.values() if not np.isnan(val)]
```
