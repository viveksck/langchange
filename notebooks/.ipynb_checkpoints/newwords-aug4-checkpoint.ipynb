```python
import scipy.sparse as sparse
from sklearn.preprocessing import normalize
import random
def get_entry_density(new_words, seq_embeds, seed=10, num_nn=5):
    reals = {"avg-cos" : np.empty((len(new_words))), "nnz" : np.empty((len(new_words))), "nn-cos" : np.empty((len(new_words)))}
    rand_reals = {"avg-cos" : np.empty((len(new_words))), "nnz" : np.empty((len(new_words))),"nn-cos" : np.empty((len(new_words))) }
    rands = {"avg-cos" : np.empty((len(new_words))), "nnz" : np.empty((len(new_words))), "nn-cos" : np.empty((len(new_words)))}
    counter = 0
    for word, year in new_words.iteritems():
        year_embed = seq_embeds.get_embed(year)
        word_embed = year_embed.represent(word)
        rand_word_embed = year_embed.represent(random.choice(year_embed.iw))
        rand_embed = sparse.rand(word_embed.shape[0], word_embed.shape[1],
                                 density=float(word_embed.nnz)/word_embed.shape[1], random_state=seed)
        normalize(rand_embed, copy=False)
        real_dist = np.sort((year_embed.m.dot(word_embed.T).data))
        reals["nn-cos"][counter] = real_dist[-num_nn-1:-2].sum()
        reals["avg-cos"][counter] = real_dist.sum()
        reals["nnz"][counter] = word_embed.nnz
        rand_real_dist = np.sort((year_embed.m.dot(rand_word_embed.T).data))
        rand_reals["nn-cos"][counter] = rand_real_dist[-num_nn-1:-2].sum()
        rand_reals["avg-cos"][counter] = rand_real_dist.sum()
        rand_reals["nnz"][counter] = rand_word_embed.nnz
        rand_dist = np.sort((year_embed.m.dot(rand_embed.T).data))
        rands["nn-cos"][counter] = rand_dist[-num_nn-1:-2].sum()
        rands["avg-cos"][counter] = rand_dist.sum()
        rands["nnz"][counter] = rand_embed.nnz
        counter +=1 
    return reals, rand_reals, rands

def get_entry_nn_density(new_words, seq_embeds, seed=10, num_nn=5):
    reals = {"avg-cos" : np.empty(num_nn*(len(new_words))), 
             "nnz" : np.empty((num_nn*len(new_words))), 
             "nn-cos" : np.empty(num_nn*(len(new_words)))}
    counter = 0
    for t_word, year in new_words.iteritems():
        nns = [entry[1] for entry in seq_embeds.get_embed(year).closest(t_word, n=num_nn+1)[1:]]
        for word in nns:
            year_embed = seq_embeds.get_embed(year)
            word_embed = year_embed.represent(word)
            real_dist = np.sort((year_embed.m.dot(word_embed.T).data))
            reals["nn-cos"][counter] = real_dist[-num_nn-1:-2].sum()
            reals["avg-cos"][counter] = real_dist.sum()
            reals["nnz"][counter] = word_embed.nnz
            counter +=1 
    return reals
```

```python
# we can also do a similar analysis but focusing on the "ambassador" nodes
def get_ams_effect(words, seq_embeds, start_year, am_end_year, end_year):
    ams = {}
    word_sims = collections.defaultdict(lambda : collections.defaultdict(lambda : float('nan')))
   # am_sims =  collections.defaultdict(lambda : collections.defaultdict(lambda : float('nan')))
    for word in words:
        ams[word] = seq_embeds.get_ambassador_list(word, start_year, am_end_year, num_a=5, first_order=False)
        if len(ams[word]) == 0:
           continue
        for year in range(start_year, end_year + 1):
            word_sims[word][year] = 0
            #am_sims[word][year] = 0
            for am in ams[word]:
                word_sims[word][year] += seq_embeds.get_embed(year).similarity(am, word)
              #  for am_2 in ams[word]:
              #      if am == am_2:
              #          continue
                 #   am_sims[word][year] += seq_embeds.get_embed(year).similarity_first_order(am, am_2)
    return word_sims
```
