```python
def get_neighborhood_con(seq_embeds, con_set, abs_set, word_set, start_year=1900, end_year=2000, num_n=10, first_order=False):
    con_scores = collections.defaultdict(lambda : collections.defaultdict(lambda : float('nan')))
    for year in range(start_year, end_year + 1):
        year_embed = seq_embeds.get_embed(year)
        for word in word_set:
            if first_order:
                neighs = year_embed.closest_first_order(word, n=num_n + 1)
            else:
                neighs = year_embed.closest(word, n=num_n + 1)
            neigh_set = set([entry[1] for entry in neighs if entry[1] != word and entry[0] != 0])
            if len(neigh_set) <= 1:
                score = float('nan')
            else:
                score = (len(neigh_set.intersection(con_set))-len(neigh_set.intersection(abs_set)))/float(len(neigh_set))
            con_scores[word][year] = score
    return con_scores
```

```python
import itertools
sns.set_style("white")
sns.set_context("talk")
def plot_coeffs(coeffs, cis, alpha=0.0001):
    colors = itertools.cycle(sns.color_palette("colorblind", n_colors=len(coeffs)))
    f, ax = plt.subplots(1,1)
    for i, term in enumerate(coeffs.index):
        color = next(colors)
        low, high = sp.stats.norm.interval(1-alpha, loc=coeffs.ix[term], scale=cis[i])
        ax.plot([low, high], [i,i], c=color, solid_capstyle="round", lw=4.5)
        ax.plot(coeffs.ix[term], i, "o", c=color, ms=8)
        ax.set_ylim(-.5, len(coeffs) - .5)
        ax.axvline(0, ls="--", c="dimgray")
        ax.set_yticks(range(len(coeffs)))
        ax.set_yticklabels(coeffs.index)
    return ax
```

```python
import matplotlib.colors
def _freedman_diaconis_bins(a):
    a = np.asarray(a)
    h = 2 * sns.iqr(a) / (len(a) ** (1 / 3))
    if h == 0:
        return np.sqrt(a.size)
    else:
        return np.ceil((a.max() - a.min()) / h)
def my_hex(x_data, y_data, gridsize=60):
    color = sns.color_palette()[0]
    color_rgb = matplotlib.colors.colorConverter.to_rgb(color)
    colors = [sns.set_hls_values(color_rgb, l=l) for l in np.linspace(1, 0, 12)]
    cmap = sns.blend_palette(colors, as_cmap=True)
    plt.hexbin(x_data, y_data, gridsize = gridsize, cmap=cmap)
```

```python
import itertools
sns.set_style("darkgrid")
sns.set_context("talk")
def plot_coeffs(coeffs, cis, alpha=0.0001):
    colors = itertools.cycle(sns.color_palette("dark", n_colors=len(coeffs)))
    f, ax = plt.subplots(1,1)
    for i, term in enumerate(coeffs.index):
        color = next(colors)
        low, high = sp.stats.norm.interval(1-alpha, loc=coeffs.ix[term], scale=cis[i])
        ax.plot([low, high], [i,i], c=color, solid_capstyle="round", lw=4.5)
        ax.plot(coeffs.ix[term], i, "o", c=color, ms=8)
        ax.set_ylim(-.5, len(coeffs) - .5)
        ax.axvline(0, ls="--", c="dimgray")
        ax.set_yticks(range(len(coeffs)))
        ax.set_yticklabels(coeffs.index)
    return ax
```

```python
def pos_interaction(cov, res, df, ax, alpha=0.0001, label_dict = {"noun" : "Noun", "adj" : "Adjective",
                                                              "verb" : "Verb", "adv" : "Adverb"}):
    low_val = np.percentile(df[cov], 10)
    high_val = np.percentile(df[cov], 90)
    slopes, stderrs = mixedmodels.get_slopes_stderrs(res)
    base = slopes[cov]
    colors = itertools.cycle(sns.color_palette("colorblind", n_colors=4))
    for param, stderr in stderrs.iteritems():
        if cov in param:
            color = next(colors)
            sepd = param.split(":")
            if len(sepd) == 1:
                pos = "noun"
                x_points, y_points = mixedmodels.simple_slope_categories(res, df, cov, None, ["verb", "adj", "adv"])
            else:
                pos = sepd[0]
                x_points, y_points = mixedmodels.simple_slope_categories(res, df, cov, pos, ["verb", "adj", "adv"])
            ci = sp.stats.norm.interval(1-alpha, scale=stderr)
            line, = ax.plot(x_points, y_points, "o", ls="-", c=color, ms=8, lw=4.5)
            line.set_label(label_dict[pos])
            ax.plot([x_points[0], x_points[0]], [y_points[0] + ci[0], y_points[0] + ci[1]], 
                    c=color, solid_capstyle="round", lw=4.5)
            ax.plot([x_points[1], x_points[1]], [y_points[1] + ci[0], y_points[1] + ci[1]], 
                    c=color, solid_capstyle="round", lw=4.5)
    return ax
```

```python
plt.rcParams["figure.figsize"] = (10, 3)
sns.set_style("whitegrid")
f, axs = plt.subplots(1, 2, sharey=True)
ax = pos_interaction("extent", pos_res8, test_df, axs[0])
ax.legend(loc="lower left")
ax.set_ylabel("Rate of Change")
ax.set_xlabel("Strength (Normalized)")
ax.set_xlim(-0.7, 2.0)
ax.text(-0.6, 0.56, "C", weight="bold", fontsize=14)
ax = pos_interaction("clust", pos_res8, test_df, axs[1])
ax.set_xlabel("Clustering (Normalized)")
ax.set_xlim(-0.9, 0.6)
ax.text(-0.85, 0.56, "D", weight="bold", fontsize=14)
plt.tight_layout()
ax.legend().remove()
plt.savefig("pos-int.pdf")
```
