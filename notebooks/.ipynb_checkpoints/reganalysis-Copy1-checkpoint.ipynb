```python
# define function that gets context concreteness score and get scores for words
def get_neighborhood_con(seq_embeds, con_set, abs_set, word_set, start_year=1900, end_year=2000, num_n=10, first_order=False):
    con_scores = collections.defaultdict(lambda : collections.defaultdict(lambda : float('nan')))
    for year in range(start_year, end_year + 1):
        year_embed = seq_embeds.get_embed(year)
        for word in word_set:
            if first_order:
                neighs = year_embed.closest_first_order(word, n=num_n + 1)
            else:
                neighs = year_embed.closest(word, n=num_n + 1)
            neigh_set = set([entry[1] for entry in neighs if entry[0] != 0])
            if len(neigh_set) <= 1:
                score = float('nan')
            else:
                score = (len(neigh_set.intersection(con_set))-len(neigh_set.intersection(abs_set)))/float(len(neigh_set))
            con_scores[word][year] = score
    return con_scores
# scores for all words per year
con_scores = get_neighborhood_con(first_orders, set(concrete_cmn), set(abs_cmn), valid_set, start_year=1900, first_order=True)
# yearly mean and stds
con_scores_ms = seriesanalysis.get_series_mean_std(con_scores, con_scores.keys())
```

```python
# plot mixed model coefficients
import itertools
sns.set_style("darkgrid")
sns.set_context("talk")
sns.set_palette("Set1")
plt.rcParams["figure.figsize"] = (7, 3)
def plot_coeffs(coeffs, cis, alpha=0.0001):
    colors = itertools.cycle(sns.color_palette("dark", n_colors=len(coeffs)))
    f, ax = plt.subplots(1,1)
    for i, term in enumerate(coeffs.index):
        color = next(colors)
        low, high = sp.stats.norm.interval(1-alpha, loc=coeffs.ix[term], scale=cis[i])
        ax.plot([low, high], [-i,-i], c=color, solid_capstyle="round", lw=4.5)
        ax.plot(coeffs.ix[term], -i, "o", c=color, ms=8)
        ax.set_ylim(-len(coeffs) + .5, 0.5)
        ax.axvline(0, ls="--", c="dimgray")
        ax.set_yticks(-1*np.arange(len(coeffs)))
        ax.set_yticklabels(coeffs.index)
    return ax
ax = plot_coeffs(full_res.fe_params[100:], np.sqrt(np.diag(full_res.cov_params()[0:full_res.k_fe]))[100:])
ax.set_yticklabels(["Frequency", "$\Delta$Frequency", "Cohesion", "Strength", "Centrality", "Concreteness"])
ax.text(0.03, -5.2, "$99.99\%$ CIs\n $p < 0.0001$", fontsize=14)
plt.tight_layout()
plt.xlabel("Marginal Effect on Rate of Semantic Change")
plt.savefig("effects.pdf")
```

```python
def get_bootstrap_k(x, y, grid, nboot=10, bootsize=10000, bwselectsize=1000):
    boot_res = []
    indices =np.random.choice(len(x), bwselectsize)
    bwreg = sm.nonparametric.KernelReg(y[indices], x[indices], var_type='c', bw='aic')
    for i in range(nboot):
        if bootsize != None:
            indices =np.random.choice(len(x), bootsize)
        else:
            indices = np.random.choice(len(x), len(x))
        kreg = sm.nonparametric.KernelReg(y[indices], x[indices], var_type='c', bw=bwreg.bw)
        res, _ = kreg.fit(data_predict=grid)
        boot_res.append(res)
    return boot_res

def get_ablated_bootstrap_k(ablated_mods, feat, nboot=10, bootsize=5000, grid=None):
    x = np.mat(noreg_train_df[feat].as_matrix()).T
    res = ablated_lmms[feat]
    y = res.model.endog - np.dot(res.model.exog, res.fe_params.as_matrix()) + np.mean(train_df["vol"])
    if grid == None:
        grid = np.linspace(np.percentile(x, 1), np.percentile(x, 99), num=100)
    return grid, get_bootstrap_k(x, y, grid, nboot=nboot, bootsize=bootsize)
```

```python
def make_bootstrap_ci(boots, alpha):
    boot_arr = np.array(boots)
    boot_means = np.mean(boots, axis=0)
    boot_lows = np.percentile(boots, alpha/2.0 * 100, axis=0)
    boot_highs = np.percentile(boots, (1-alpha/2.0) * 100, axis=0)
    return boot_means, boot_highs, boot_lows
```
